```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
import numpy as np
import scipy as sp
#import scipy.constnats as const
```

```python
x = np.linspace(0, 2*np.pi, num=300)

plt.plot(x, np.sin(x), lw=4, label="Theory")
plt.plot(x, np.sin(x - 0.3), lw=4, label="Experiment")

plt.legend()
plt.xlabel("Polarization Angle")
plt.ylabel("Rotation speed")
plt.margins(x=0)
plt.show()
```
